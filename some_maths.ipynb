```python
import sympy as sp
```

```python
p1, p2, p3, p4, g1, g2, g3, b = sp.symbols('p1 p2 p3 p4 g1 g2 g3 b')
Lp = sp.Matrix([
    [p1, 0, b, 0],
    [g1, p2, 0, 0],
    [0, g2, p3, 0],
    [0, 0, g3, p4]
])
eigen= Lp.eigenvects()
```

```python
r1_func = sp.lambdify([(p1, p2, p3, p4, g1, g2, g3, b)], sp.simplify(eigen[1][0]), 'numpy')
```

3. Теперь посмотрим как обстоит дело на эксперименте, учитывая перемещения 

Для удобства вычислений на каждом такте для каждой страны будем искать классическую матрицу Лесли

```python
df5 = pd.read_csv('experiment_2_d.txt', names=['num', 'country', 'age', 'i'])
df5['country'] = df5['country'].astype('str')
df5['age'] = df5['age'].astype(int)
df5['num'] = df5['num'].astype(int)
df5['i'] = df5['i'].astype(int)
```

```python
LL = {}
countries = df5['country'].drop_duplicates().to_list()
demog = {}
```

```python
for t in countries:
    LL[t] = np.zeros(shape=(50, 4, 4))
    demog[t] = {}
    for a in range(4):
        demog[t][a] = df5[(df5['country'] == t) & (df5['age'] == a)]['num'].to_list()
```

```python
for i in range(49):
    for t in countries:
        LL[t][i][0][2] = demog[t][0][i+1] / demog[t][2][i] 
        LL[t][i][1][0] = demog[t][1][i+1] / demog[t][0][i] 
        LL[t][i][2][1] = demog[t][2][i+1] / demog[t][1][i]
        LL[t][i][3][2] = demog[t][3][i+1] / demog[t][2][i]
```

```python
eigenvalues = {}
for t in countries:
    eigenvalues[t] = [abs(max(np.linalg.eig(LL[t][i])[0])) for i in range(50)] 
```

```python
import plotly.graph_objects as go
```

Все собственные значения вещественные, поэтому с чистой совестью строю двумерный график и беру модули

```python
fig = go.Figure()
for label, data in eigenvalues.items():
    fig.add_trace(go.Scatter(x=list(range(1, len(data))), y=data[:-1], mode='lines+markers', name=label))
fig.add_trace(go.Scatter(x=list(range(1, len(data))), y=[max(np.abs(eigenvalues2))] * 49, mode='lines+markers', name='реальная + выживание - катания'))
fig.update_layout(title='График собственных значений для классической модели Лесли', xaxis_title='такт', yaxis_title='Собственные значения')
fig.show()
```
